```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Lead the datasets 
iris = load_iris()
X = pd.DataFrame(iris.data, columns = iris.feature_names) # Feature X 
y = iris.target
```

```python
# Split 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.44, random_state = 42, stratify = y
)
print(f"Training set size: {len(X_train)} samples")
print(f"Test set size: {len(X_test)} samples")
```

```python
# Create and Train the model
clf =  DecisionTreeClassifier(
    criterion = 'gini',
    max_depth = 3,
    random_state = 42
)
clf.fit(X_train, y_train)
```

```python
# Prediction
y_pred = clf.predict(X_test)
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy:.4f}")
```

```python
# Visualize
plt.figure(figsize=(15,10))
plot_tree(
    clf,
    class_names = iris.target_names,
    feature_names = iris.feature_names,
    filled = True,
    rounded = True,
    fontsize = 10
)
plt.title("Decission Tree Classifier for Iris dataset.(Max Depth = 3)")
plt.show()
```

```python
# Prediction
X_new = np.array([
    [6.0, 3.5, 5.5, 1.2]
])
new_prediction = clf.predict(X_new)

# Map numerial prediction with Species name
iris_target_name = np.array(['Setosa', 'Versicolor', 'Virginica'])
predicted_species = iris_target_name[new_prediction]
print(f"Predicted Species Name: {predicted_species}")
```
