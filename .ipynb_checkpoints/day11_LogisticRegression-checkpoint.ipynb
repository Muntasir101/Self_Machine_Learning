# Visualize

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# -------------------------------------------------
# 1. Data
# -------------------------------------------------
X = np.array([[1], [3], [5], [7]])      # study hours
y = np.array([0, 0, 1, 1])              # 0 = fail, 1 = pass

# -------------------------------------------------
# 2. Train logistic regression
# -------------------------------------------------
model = LogisticRegression()
model.fit(X, y)

# -------------------------------------------------
# 3. Predictions for the *original* points
# -------------------------------------------------
y_pred_class = model.predict(X)                 # 0/1
y_pred_prob  = model.predict_proba(X)[:, 1]     # probability of class 1

# -------------------------------------------------
# 4. Smooth sigmoid curve (for visualisation)
# -------------------------------------------------
X_curve = np.linspace(0, 8, 300).reshape(-1, 1)
probs   = model.predict_proba(X_curve)[:, 1]

# -------------------------------------------------
# 5. Plot
# -------------------------------------------------
plt.figure(figsize=(8, 6))

# 5a – Actual outcomes (0/1)
plt.scatter(X, y,
            color='blue', s=100, edgecolor='k',
            label='Actual (0=Fail, 1=Pass)', zorder=5)


# 5c – Sigmoid probability curve
plt.plot(X_curve, probs, color='red', linewidth=2,
         label='Predicted probability (sigmoid)')

# 5d – Decision threshold
plt.axhline(y=0.5, color='green', linestyle='--',
            label='Threshold p=0.5')


# -------------------------------------------------
# 6. Formatting
# -------------------------------------------------
plt.xlabel('Study hours')
plt.ylabel('Outcome / Probability')
plt.title('Logistic Regression – Actual vs Predicted')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-0.1, 1.2)
plt.xlim(0, 8)
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 7. Print table (same as you had)
# -------------------------------------------------
print("\nHours | Actual | Pred. Prob")
print("-" * 30)
for hours, actual, prob in zip(X.flatten(), y, y_pred_prob):
    print(f"{hours:5} | {actual:6} | {prob:.3f}")

# -------------------------------------------------
# 8. Example prediction for 4 hours (as you asked)
# -------------------------------------------------
pred_4h = model.predict([[4]])[0]
prob_4h = model.predict_proba([[4]])[:, 1][0]
print(f"\nPrediction for 4 study hours -> class: {pred_4h}, prob: {prob_4h:.3f}")
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve,auc, confusion_matrix

# 1. Data
# -------------------------------------------------
X = np.array([[1],[2], [3], [5], [7], [8]])      # study hours
y = np.array([0, 0, 0, 1, 1, 1]) 

# 2. Train logistic regression
# -------------------------------------------------
model = LogisticRegression()
model.fit(X, y)

# predictions and probabilities
y_pred = model.predict(X)
y_probs = model.predict_proba(X)[:,1]

# Calculate metrics
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)

print(accuracy)
print(precision)
print(recall)
print(f1)

confusion_metrics = confusion_matrix(y, y_pred)
print(confusion_metrics)
```
