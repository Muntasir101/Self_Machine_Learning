```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge 
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import warnings 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
# Sample dataset
# # Perfect polynomial relationship
df['Price'] = 50*df['Size'] + 10000*df['Bedrooms'] - 2000*df['Age'] + 0.5*(df['Size']**2)/1000
data = {
    'Size': [1500,1800,2400,2000,1700,2200,1600,1900,2100,2300],
    'Bedrooms' :[3,4,2,4,3,2,5,3,2,3],
    'Age': [10,4,6,2,8,7,15,20,25,9],
    'Price': [328500,416600,582400,500000,381450,547300,411800,454950,516050,566350]
}
```

```python
# Create dataframe
df = pd.DataFrame(data)
```

```python
# Independent variables(Size,Bedrooms,Age)
X = df[['Size', 'Bedrooms', 'Age']]
# Depepdent variable
y = df['Price']
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=42)
```

```python
# Model 1: Linear Regression
linear_model =  LinearRegression()
linear_model.fit(X_train,y_train)
y_test_pred_linear = linear_model.predict(X_test)
```

```python
# Model 2: Polynomial Regression
poly_model = make_pipeline(
    StandardScaler(),
    PolynomialFeatures(degree = 1, include_bias=False),
    Ridge(alpha = 1.0) # Regularization to avoid overfitting
)
```

```python
poly_model.fit(X_train, y_train)
y_test_pred_poly = poly_model.predict(X_test)
```

```python
print(r2_score(y_test, y_test_pred_linear))
print(y_test, y_test_pred_linear)
```

```python
#prediction for new house
new_house = np.array([[2000, 4, 2]])
pred_linear = linear_model.predict(new_house)[0]
pred_poly = poly_model.predict(new_house)[0]
print(pred_linear)
print(pred_poly)
```

```python
# visualize
plt.figure(figsize=(10,5))
```

```python
# Subplot 1: Linear
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_test_pred_linear, color = 'blue', alpha = 0.7)
plt.plot([y_test.min(),y_test.max()], [y_test.min(),y_test.max()], 'r--')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Linear Model")
```

```python
# Subplot 2: Polynomial
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_test_pred_poly, color = 'green', alpha = 0.7)
plt.plot([y_test.min(),y_test.max()], [y_test.min(),y_test.max()], 'r--')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Polynomial Model")
plt.tight_layout()
plt.show()
```

```python
# pairplot
sns.pairplot(df,kind='reg', diag_kind ='kde', plot_kws={'line_kws':{'color':'red'}})
plt.suptitle("pairplot with Regression Lines", y=1.02)
plt.show()
```

```python
# Train score for Polynomial
poly_train_r2 = r2_score(y_train, poly_model.predict(X_train))
print(poly_train_r2)
```

```python
# Get actual r2 score for Polynomial model
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(poly_model, X, y, cv = 5, scoring='r2')
print(cv_scores.mean())
print(cv_scores.std())
```

```python
# Train score for Linear
linear_train_r2 = r2_score(y_train, linear_model.predict(X_train))
print(linear_train_r2)
```

```python
# Get actual r2 score for Linear model
cv_scores = cross_val_score(linear_model, X, y, cv = 5, scoring='r2')
print(cv_scores.mean())
print(cv_scores.std())
```
