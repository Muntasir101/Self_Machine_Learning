```python
# Basic Import 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet 
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.datasets import make_regression
import warnings 
warnings.filterwarnings('ignore')
print("all Libraries imported successfully")
```

# Simple Linear Regression - House Price Predeiction

```python
# create sample dataset
np.random.seed(42)
square_footage = np.random.normal(1500,500,100)
house_prices = 50000 + 200 * square_footage + np.random.normal(0, 30000, 100)

# create dataframe
house_data = pd.DataFrame({
    'Square_Footage':square_footage,
    'House_Prices': house_prices
                         })
print("House Data Sample:")
print(house_data.head())
```

```python
# Visualize
plt.figure(figsize=(10,6))
plt.scatter(house_data['Square_Footage'],house_data['House_Prices'], alpha=0.7)
plt.xlabel('Square Footage')
plt.ylabel('House_Prices')
plt.title('House Price vs Square Footage')
plt.grid(True,alpha=0.3)
plt.show()
```

```python
# Implement Simple Linear Regression 
def simple_linear_regression(x,y):
    # Calculate Mean
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # Calaculate coeffecient using OLS formula 
    numerator = np.sum((x - x_mean) * (y- y_mean))
    denominator = np.sum((x - x_mean) **2)
    beta_1 = numerator/denominator  # sloop 
    beta_0 = y_mean - beta_1 * x_mean  # intercept

    return beta_0, beta_1

# Calculate coefficient manually
x = house_data['Square_Footage'].values
y = house_data['House_Prices'].values
beta_0, beta_1 = simple_linear_regression(x,y)

print(f'Intercept: {beta_0:.2f}') 
print(f'Slope: {beta_1:.2f}') 
print(f'Equation Price = {beta_0:.2f} + {beta_1:.2f}*Square_Footage')
```

```python
# Compare with Scikit-Learn
# prepare data
X = house_data[['Square_Footage']]
y = house_data['House_Prices']

# Create and Train model
model = LinearRegression()
model.fit(X,y)

print(f"\nScikit-Learn Results:")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Scope: {model.coef_[0]:.2f}")
```

```python
# visualize the Regression Line
plt.figure(figsize = (10,6))
plt.scatter(house_data['Square_Footage'],house_data['House_Prices'], alpha=0.7, label='Actual data')

# Generate predcitions for the line
x_line = np.linspace(house_data['Square_Footage'].min(), house_data['Square_Footage'].max(), 100)
y_line = model.predict(x_line.reshape(-1,1))
plt.plot(x_line,y_line,color='red', linewidth = 2, label='Regression Line')
plt.xlabel('Square Footage')
plt.ylabel('House_Prices')
plt.title('Linear Regression: House Price Prediction')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
