# 1. Basic KNN Implementation

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Load the datase
iris =  load_iris()
X = iris.data  # Features
y = iris.target  # Target 
```

```python
# create DataFrame for better visualize
df = pd.DataFrame(X, columns = iris.feature_names)
df['species'] = y
df['species_name'] = [iris.target_names[i] for i in y] 
```

```python
df
```

```python
print('Dataset Overview')
print(df.head())
print(df.shape)
print(f"Species distribution:\n {df['species_name'].value_counts()}")
```

# 2. Data Exploration

```python
# Visualize
plt.figure(figsize = (15,10))

# pairplot
sns.pairplot(df, hue ='species_name', palette = 'viridis')
plt.suptitle('Iris Dataset - Feature Relationship', y=1.02)
plt.show()
```

# 3. Train-Test Split and Model Training

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

```python
# Initilize and Train KNN
knn = KNeighborsClassifier(n_neighbors=3) # k=3
knn.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"\nKNN Model Performance (k=3):")
print(f"Accuracy {accuracy:.4f}")
```

# 4. Finding the optimal K value

```python
# Test different k value
k_values = range(1, 21)
train_scores = []
test_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test, y_test))

# plot accuracy vs k value
plt.figure(figsize = (12,6))
plt.plot(k_values, train_scores, 'o-', label='Training Accuracy', alpha = 0.7)
plt.plot(k_values, test_scores, 'o-', label='Test Accuracy', alpha = 0.7)
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.title('KNN: Accuracy vs k Value')
plt.legend()
plt.grid(True, alpha = 0.3)
plt.xticks(k_values)
plt.show()

print(f"All test scores: {test_scores}")

# Find the optimal k
optimal_k = k_values[np.argmax(test_scores)]
print(f"Optimal k value: {optimal_k} with test accuracy: {max(test_scores):.4f}")
```

```python
# Final Model with optimal k
optimal_knn = KNeighborsClassifier(n_neighbors=optimal_k)
optimal_knn.fit(X_train, y_train)

# Final prediction
y_pred_optimal = optimal_knn.predict(X_test)

print("Final Model Performance")
print(f"Accuracy: {accuracy_score(y_test, y_pred_optimal):.4f}")
print(f"Optimal K: {optimal_k}")
```

# 5. Making Prediction on New Data

```python
def predict_new_flower(sepal_length, sepal_width, petal_length, petal_width):
    new_flower = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    prediction = optimal_knn.predict(new_flower)
    probability = optimal_knn.predict_proba(new_flower)

    species = iris.target_names[prediction[0]]
    print(f"Predicted Species: {species}")
    print("Prababilities")

    for i, prob in enumerate(probability[0]):
        print(f" {iris.target_names[i]}: {prob:.4f}")

    return species

# Test with some examples values
print("Example Predictions:")
print("\nExample 1:")
predict_new_flower(5.1, 3.5, 1.4, 0.2)

print("\nExample 2:")
predict_new_flower(6.1, 4.5, 3.4, 2.2)

print("\nExample 2:")
predict_new_flower(0.1, 5.0, 5.7, 5.2)
```
