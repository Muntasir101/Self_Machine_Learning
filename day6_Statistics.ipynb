```python
# Binomial Distribution
```

```python
from scipy.stats import binom
```

```python
# parameters
n = 10   # Number of trails
p = 0.5  # orobability of success
```

```python
# Possible outcomes
x = np.arange(0,n+1)
print(x)
```

```python
# probability mass function(PMF)
pmf = binom.pmf(x,n,p)
print(pmf)
```

```python
#plot
plt.bar(x,pmf)
plt.xlabel("Number of Success")
plt.ylabel("Probability")
plt.title("Binomial Distribution")
plt.show()
```

```python
# Poisson Distribution
```

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
# parameter
lamda = 3   # average number of events
```

```python
# possible outcomes
x = np.arange(0,10)
print(x)
```

```python
# probability mass function(PMF)
pmf = poisson.pmf(x,lamda)
print(pmf)
```

```python
#plot
plt.bar(x,pmf)
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Poisson Distribution")
plt.show()
```

```python
# Example data
hours = np.array([2,4,6,10,20])
scores = np.array([20,30,50,70,90])
```

```python
# Covariance 
cov_matrix = np.cov(hours,scores)
print(cov_matrix)
cov_matrix.shape
```

```python
# Covariance between hours and scores
cov_hours_scores = cov_matrix[0,1]
print(cov_hours_scores)
```

```python
# Correlation
```

```python
corr_matrix=np.corrcoef(hours,scores)
print(corr_matrix)
```

```python
# Correlation coefficient
corr_hours_scores = corr_matrix[0,1]
print(corr_hours_scores)
```

```python
# Scenario: 
# A factory produce light bulbs, and each bulb has 5% chance of being defective.
# If a batch contains 20 bulbs, find 
# 1. probability that exactly 2 bulbs are defective 
# 2. probability that at most 1 bulb is defective
```

```python
from scipy.stats import binom
```

```python
# parameters
n = 20   # trails
p = 0.05  # porobability of defects
```

```python
# Exactlt e defective
prob_2 = binom.pmf(2,n,p)
print("P(exactly 2 defective) = ",prob_2)
```

```python
# at most 1 bulb is defective
prob_at_most_1 = binom.cdf(1,n,p)
print("P(at most 1 defective) = ",prob_at_most_1)
```
