```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Sample dataset
data = {
    'Size': [1500,1800,2400,2000,1700,2200,1600,1900,2100,2300],
    'Bedrooms' :[3,4,2,4,3,2,5,3,2,3],
    'Age': [10,4,6,2,8,7,15,20,25,9],
    'Price': [300000,350000,400000,320000,380000,360000,280000,250000,230000,310000]
}
```

```python
# Create dataframe
df = pd.DataFrame(data)
```

```python
df
```

```python
# Independent variables(Size,Bedrooms,Age)
X = df[['Size', 'Bedrooms', 'Age']]
# Depepdent variable
y = df['Price']
```

```python
# Create and Train the model
model =  LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
print("Coefficients:",model.coef_)
print("Intercept:",model.intercept_)
print("Mean Squared Error", mean_squared_error(y,y_pred))
print("R^2 Score:", r2_score(y,y_pred))
```

```python
#prediction for new house
new_house = np.array([[2000, 3, 10]])
predicted_price = model.predict(new_house)
print("Predicted Price:", predicted_price[0])
```
