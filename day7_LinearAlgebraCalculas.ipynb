## VECTORS - Representing of Data
- A Vector is an ordered list of numbers.
- It represents data points, weights of features.

x = [1,2,3] 

in ML each vectors ofter represents:
- A Feature vector(input)
- A Weighted vector(model parameters)

## MATRICES - Data Collections 
- A Matrix is a collection of vectors(rows or columns)

## DOT PRODUCT - Measuring Alignment
- The Dot product between two vectors measures similarity or alignment.

a.b = a1b1 + a2b2+....+anbn

[1,2,3].[4,5,6] = 1*4 + 2*5 + 3* 6 = 32 

## NORMS - Length or Magnitude 
- The norm of a vector = its length.

1. L2 Norm(Euclidean)
2. L1 Norm(Manhattan)
  

## Vectors Example

- Let's say you have a student's data:

1. Study hous = 2
2. IQ(scaled) = 3
3. Experiances = 5

vector x = [2,3,5]

```python
import numpy as np
x = np.array([2,3,5])
print("Vector x:", x)
print("Shape:", x.shape) # 1D 
```

```python
# [2,3,5], [4,5,6]
```

```python
x = np.array([[2,3,5],[4,5,6]])
print("Vector x:", x)
print("Shape:", x.shape)  # 2D
```

```python
# Dot product - How similar or Aligned two vectors are
a = np.array([2,3,5])
b = np.array([4,5,6])
dot_product = np.dot(a,b)
print("Dot Product:", dot_product)
```

# ML Meaning
- in Linear Regression:
  y = w.x + b

w = weights(model parameters)
x = input vector 
b = bias

```python
w = np.array([1.5, 0.5, 2.0])   # study hour/features
x = np.array([20, 3, 50])   # IQ/weights
b = 1.0  
y = np.dot(w,x) + b
print("Predicted value:", y)
```

```python
# Norms
x = np.array([3,4])
l2_norm = np.linalg.norm(x)   # Eclidean: root(x1 squar + x2 square) 
l1_norm = np.linalg.norm(x,1)  # Manhattan : x1+x2
print(l2_norm)
print(l1_norm)
```

# House Price Prediction(Regression)
## Problem: Predict the price of a house bases on features like:
- Area(sq ft)
- Bedrooms
- Bathrooms
- Location Score

```python
x = np.array([2000, 3, 2, 8])  # house features
w = np.array([50, 10, 5, 20])  # model wights 
b = 100                        # bias/base price
predcited_price = np.dot(w,x)+b
print("Predecited Price:",predcited_price)
```

# ●	Class 2: Matrix operations in NumPy (inverses, eigenvalues)

```python
A = np.array([[1,2],[3,4]])
print(A.T)
```

```python
# inverse
A = np.array([[1,2],[3,4]])
A_inv = np.linalg.inv(A)
print(A_inv)
print(np.dot(A, A_inv))  # should be identity
```

```python
# eigenvalues
eigvals, eigvecs = np.linalg.eig(A)
print("eigvals:", eigvals)
print("eigvecs:", eigvecs)
```

```python
# Derivatives
import sympy as sp
x = sp.symbols('x')
f = x**2 
df = sp.diff(f,x)
print("Derivatives",df)
print("Value at x=3:", df.subs(x,3)) # how much to adjust parameters to reduce error. 
```

```python
# Partial Derivatives
x, y = sp.symbols('x y')
f = x**2 + y**2
df_dx = sp.diff(f,x)
df_dy = sp.diff(f,y)
print("df/dx: ",df_dx)
print("df/dy: ", df_dy)
```
